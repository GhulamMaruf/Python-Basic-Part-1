# Condition: if, elif, else


<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Equals a == b </p>
<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Not Equals a != b </p>
<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Greatar than b, a > b </p>
<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Greatar than b or equal, a >= b </p>
<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Less than a, a < b </p>
<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">Less than a or equal, a <= b </p>

```python
_large_val = 50
_small_val = 30

if _large_val > _small_val:
    print('You are correct')
elif _large_val < _small_val:
    print('You are wrong')
else:
    print('You are nothing')
```

```python
_large_num = 400
_small_num = 30



if (_large_num > _small_num):
    print('Go ahead')
    if (_large_num >= _small_num):
        print('You are on the correct road')
    else:
        print('You have to study more')

elif (_large_num < _small_num):
    print('You wrote small number in the place of large number')
else:
    print('You are undone! Start another session ')
```

```python
for y in range(1,13,3):
    print(y)
```

<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">If there need to add all, keep a variable with value of zero(0)</p>

```python
xy = [5,10,2]

total_result = 1

for x in xy:
    total_result = total_result * x
    print(total_result)
    
print("Total result is: ",total_result)
```

# while loop

```python
x = 1

while x < 10:
    print(x)
    x = x + 1
```

<p style="color: #E58100; font-size: 22px; font-weight: bold;font-family: 'Comfortaa', cursive;">In while loop, first declare a starting value, then while loop condition, then print, then increment or decrement</p>

```python
y = 20

while y < 40:
    print(y)
    y+=1
```

```python
y = 5

while y <= 20:
    print(y)
    y+=1
```

```python
y = 10

while y > 5:
    print(y)
    y-=1
```

```python
u = 50

while u > 5:
    print(u)
    
    u -= 5
```
